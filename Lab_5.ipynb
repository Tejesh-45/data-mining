```python
import pandas as pd

df = pd.read_csv('/content/employee_data (1).csv')
print("Original Dataset:")
print(df)

df.isna()

df.isnull().sum()

df['Age'].mean()

df['Salary'].mean()

df['Job_Satisfaction'].mean()

df['Job_Satisfaction'].mode()

df['Job_Satisfaction'].median()

df['Age'].fillna(df['Age'].mean())

df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Salary'] =df['Salary'].fillna(df['Salary'].mean())

df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mean())

df

print("Dataset after handling missing values:")
print(df)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[[' Work_Hours_Per_Week', 'Job_Satisfaction']] = scaler.fit_transform(df[['Work_Hours_Per_Week', 'Job_Satisfaction']])

df['Age'] = (df['Age'] - df['Age'].min()) / df['Age'].max() - df['Age'].min()

from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()

df[['Salary']] = s_scaler.fit_transform(df[[ 'Salary']])

df['Salary']

df['Job_Satisfaction'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()

df['Job_Satisfaction']
```
