```python
import pandas as pd


df = pd.read_csv('/content/dataset.csv', delimiter=';')

df.describe()


```

```python
df.info()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))
```

```python
plt.subplot(2, 3, 1)
```

```python
plt.plot(df['fixed acidity'], df['volatile acidity'], 'b-o', alpha=0.6)
plt.subplot(2, 3, 1)
plt.plot(df['alcohol'], df['quality'], 'b-o', alpha=0.6)
plt.title('Line Plot: Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
```

```python
import seaborn as sns
plt.subplot(2, 3, 2)
sns.scatterplot(x='residual sugar', y='chlorides', data=df, hue='quality', palette='viridis')
plt.title('Scatter Plot: Residual Sugar vs Chlorides')
```

```python
plt.subplot(2, 3, 3)
sns.barplot(x='quality', y='alcohol', data=df, ci=None, palette='pastel')
plt.title('Bar Plot: Average Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
```

```python
plt.subplot(2, 3, 4)
sns.histplot(df['pH'], bins=20, kde=True, color='purple')
plt.title('Histogram: pH Distribution')
plt.xlabel('pH')
plt.ylabel('Frequency')
```

```python
plt.subplot(2, 3, 5)
quality_counts = df['quality'].value_counts()
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart: Quality Distribution')
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap: Correlation Matrix')
```

```python
plt.tight_layout()
plt.show()
```
